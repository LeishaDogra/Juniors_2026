1.1 Exercise 1: FizzBuzz Implementation

```python
for i in range(1, 101) :
    if i % 3 == 0 and i % 5 == 0 :
        print("FizzBuzz")
    elif i % 3 == 0 :
        print("Fizz")
    elif i % 5 == 0 :
        print("Buzz")
    else :
        print(i)
```

1.2 Exercise 2: Prime Numbers within a Range


```python
a = int(input("Enter the smaller number: "))
b = int(input("Enter the larger number: "))

def prime(n) :
    for i in range(2, int(n**0.5) + 1) :
        if n % i == 0 :
            return False
    return True

for i in range(a, b + 1) :
    if prime(i) :
        print(i)
```

2.1 Exercise 3: Matrix Operations

```python
import numpy as np
m1 = np.random.randint(1, 11, size=(3, 3))
m2 = np.random.randint(1, 11, size=(3, 3))

print("Matrix 1:\n", m1)
print("Matrix 2:\n", m2)

mAdd = m1 + m2
print("Addition of matrices A and B:\n", mAdd)

mMult = m1 * m2
print("Element-wise multiplication of A and B:\n", mMult) 

mDot = np.dot(m1, m2)
print("Dot product of A and B:\n", mDot)

mDet = int(np.linalg.det(m1))
print("Determinant of matrix A:\n", mDet)
```

2.2 Exercise 4: Statistical Analysis


```python
import numpy as np 

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
median = np.median(data)
standard_deviation = np.std(data)
variance = np.var(data)
percentile25 = np.percentile(data, 25)
percentile75 = np.percentile(data, 75)

print("Mean:", round(mean, 2))
print("Median:", round(median, 2))
print("Standard Deviation:", round(standard_deviation, 2))
print("Variance:", round(variance, 2))
print("25th Percentile:", round(percentile25, 2))
print("75th Percentile:", round(percentile75, 2))
```
